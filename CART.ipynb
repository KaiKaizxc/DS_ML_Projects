```python
from sklearn.datasets import load_iris 

from sklearn.tree import DecisionTreeClassifier 

```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
iris["data"][:5]
```

```python
iris["target"]
```

```python
iris["target_names"]
```

```python
iris["feature_names"]
```

```python
X = iris.data[:, 2:]
y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth = 2)
tree_clf.fit(X, y)
```

```python
import matplotlib.pyplot as plt 
from sklearn import tree


fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(tree_clf, 
                   feature_names=iris["feature_names"][2:],  
                   class_names=iris["target_names"],
                   filled=True)
```

```python
tree_clf.predict_proba([[5, 1.5]])
```

```python
tree_clf.predict([[5, 1.5]])
```

```python
# how does it decide how to split and split based on which number? 
# it uses GINI impurity, and the largest impurity is the most "impure"
# so it uses the lowest one 
# gini impurity = 1 - (probs of success)^2 - (probs of failure)^2 
# then use weighted of each leaf nodes to calculate gini impurity of parent 
# then choose the variable with the split point which has the lowest gini impurity
```

```python
# Decision Tree REGRESSOR 
```

```python
from sklearn.tree import DecisionTreeRegressor 

tree_reg = DecisionTreeRegressor(max_depth = 2)
tree_reg.fit(X, y)

```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(tree_reg, 
                   feature_names=iris["feature_names"][2:],  
                   class_names=iris["target_names"],
                   filled=True)
```
